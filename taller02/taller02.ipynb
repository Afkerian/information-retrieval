## Parte 1: Preparación de Datos
### Creación de la Matriz de Términos-Documento
Cada grupo de estudiantes deberá crear una tabla con las categorías de palabras como filas y los textos como columnas. Para cada texto, los estudiantes contarán cuántas veces aparecen palabras pertenecientes a cada categoría y llenarán la tabla con estos conteos.

```python
import os
import re

def read_data(archivo):
    with open(archivo, 'r', encoding='utf-8') as f:
        texto = f.read()
        palabras = re.findall(r'\b[a-zA-ZáéíóúÁÉÍÓÚñÑ]+\b', texto.lower())  # Solo palabras con letras y caracteres acentuados
    return set(palabras)


def palabras_en_carpeta(ruta_carpeta):
    todas_palabras = set()
    for archivo in os.listdir(ruta_carpeta):
        if archivo.endswith('.txt'):
            ruta_completa = os.path.join(ruta_carpeta, archivo)
            palabras_archivo = read_data(ruta_completa)
            todas_palabras.update(palabras_archivo)
    return todas_palabras

carpeta_data = 'data/'
palabras_totales = palabras_en_carpeta(carpeta_data)
print('Total de Palabras: ',len(palabras_totales))
```

```python
import os
import pandas as pd
import re

def leer_textos_en_carpeta(ruta_carpeta):
    textos = []
    for archivo in os.listdir(ruta_carpeta):
        if archivo.endswith('.txt'):
            ruta_completa = os.path.join(ruta_carpeta, archivo)
            with open(ruta_completa, 'r', encoding='utf-8') as f:
                texto = f.read()
                textos.append(texto)
    return textos

def obtener_palabras_unicas(textos):
    todas_palabras = set()
    for texto in textos:
        palabras = re.findall(r'\b[a-zA-ZáéíóúÁÉÍÓÚñÑ]+\b', texto.lower())
        todas_palabras.update(palabras)
    return list(todas_palabras)

def contar_frecuencia_palabras(texto, palabras_unicas):
    frecuencia_palabras = {palabra: texto.lower().count(palabra) for palabra in palabras_unicas}
    return frecuencia_palabras

carpeta_data = 'data'
textos = leer_textos_en_carpeta(carpeta_data)
palabras_unicas = obtener_palabras_unicas(textos)

filas = []
for texto in textos:
    frecuencia = contar_frecuencia_palabras(texto, palabras_unicas)
    filas.append(frecuencia)

df = pd.DataFrame(filas, columns=palabras_unicas)
#print(df)
df.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df_normalized.head(20)
```

```python
# saving the dataframe
df.to_csv('dataframes/file1.csv')

df_normalized.to_csv('dataframes/file1_normalized.csv')
```

```python
# Crear un DataFrame con todas las palabras únicas como columnas
df_comparacion = pd.DataFrame(columns=palabras_unicas)

# Iterar sobre cada consulta y contar cuántas veces aparece cada palabra
for i, consulta in enumerate(query):
    fila = [consulta.count(palabra) for palabra in palabras_unicas]
    df_comparacion.loc[i] = fila

df_comparacion.head(20)
```
