## Parte 1: Preparación de Datos
### Creación de la Matriz de Términos-Documento
Cada grupo de estudiantes deberá crear una tabla con las categorías de palabras como filas y los textos como columnas. Para cada texto, los estudiantes contarán cuántas veces aparecen palabras pertenecientes a cada categoría y llenarán la tabla con estos conteos.

```python
import os
import re

def read_data(archivo):
    with open(archivo, 'r', encoding='utf-8') as f:
        texto = f.read()
        palabras = re.findall(r'\b[a-zA-ZáéíóúÁÉÍÓÚñÑ]+\b', texto.lower())  # Solo palabras con letras y caracteres acentuados
    return set(palabras)


def palabras_en_carpeta(ruta_carpeta):
    todas_palabras = set()
    for archivo in os.listdir(ruta_carpeta):
        if archivo.endswith('.txt'):
            ruta_completa = os.path.join(ruta_carpeta, archivo)
            palabras_archivo = read_data(ruta_completa)
            todas_palabras.update(palabras_archivo)
    return todas_palabras

carpeta_data = 'data/'
palabras_totales = palabras_en_carpeta(carpeta_data)
print('Total de Palabras: ',len(palabras_totales))
```

```python
import os
import pandas as pd
import re

def leer_textos_en_carpeta(ruta_carpeta):
    textos = []
    for archivo in os.listdir(ruta_carpeta):
        if archivo.endswith('.txt'):
            ruta_completa = os.path.join(ruta_carpeta, archivo)
            with open(ruta_completa, 'r', encoding='utf-8') as f:
                texto = f.read()
                textos.append(texto)
    return textos

def obtener_palabras_unicas(textos):
    todas_palabras = set()
    for texto in textos:
        palabras = re.findall(r'\b[a-zA-ZáéíóúÁÉÍÓÚñÑ]+\b', texto.lower())
        todas_palabras.update(palabras)
    return list(todas_palabras)

def contar_frecuencia_palabras(texto, palabras_unicas):
    frecuencia_palabras = {palabra: texto.lower().count(palabra) for palabra in palabras_unicas}
    return frecuencia_palabras

carpeta_data = 'data'
textos = leer_textos_en_carpeta(carpeta_data)
palabras_unicas = obtener_palabras_unicas(textos)

filas = []
for texto in textos:
    frecuencia = contar_frecuencia_palabras(texto, palabras_unicas)
    filas.append(frecuencia)

df = pd.DataFrame(filas, columns=palabras_unicas)
#print(df)
df.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

df_normalized.head(20)
```

```python
# saving the dataframe
df.to_csv('dataframes/file1.csv')

df_normalized.to_csv('dataframes/file1_normalized.csv')
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Suponiendo que df_corpus y df_query son tus dataframes
# Calcular la distancia coseno para cada fila de df_query con respecto a cada fila de df_corpus

distances = []
for index1, row1 in df_query.iterrows():
    row_distances = []
    for index2, row2 in df_corpus.iterrows():
        similarity = cosine_similarity([row1], [row2])[0][0]
        row_distances.append(similarity)
    distances.append(row_distances)

# Crear un nuevo dataframe con las distancias
distances_df = pd.DataFrame(distances)

print(distances_df)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Definir las consultas
query = [
    "Historia medieval de las ciudades europeas",
    "Principales destinos turísticos en Europa",
    "Influencia de la realeza en la cultura europea",
    "Importancia de los parques urbanos en las ciudades",
    "Gastronomía típica en capitales europeas",
    "Eventos deportivos icónicos en Europa",
    "Celebraciones de Navidad en ciudades europeas",
    "Museos de arte importantes en Europa",
    "Efectos de la Segunda Guerra Mundial en ciudades europeas",
    "Arquitectura histórica en ciudades europeas",
    "Clubes de fútbol famosos y sus estadios en Europa",
    "Transformación urbana y regeneración de espacios",
    "Tradición cervecera en ciudades europeas",
    "Mercados y comercio tradicional en Europa",
    "Centros de arte y creatividad en ciudades modernas",
    "Actividades de ocio y entretenimiento en ciudades metropolitanas",
    "Conservación de la naturaleza y la vida silvestre en zonas urbanas",
    "Planificación de eventos culturales y festivales en ciudades",
    "Desarrollo del transporte y la infraestructura urbana",
    "Impacto de la tecnología en la vida urbana"
]

# Crear DataFrame de las consultas
df_query = pd.DataFrame(query, columns=['Texto'])

# Calcular similitud para cada consulta con respecto a los corpus
similarity_scores = []
for index, row in df_query.iterrows():
    similarity_scores.append(distances_df.iloc[index].tolist())

# Crear DataFrame con los resultados
result_df = pd.DataFrame(similarity_scores, columns=[f'Corpus_{i+1}' for i in range(len(similarity_scores))])
result_df.insert(0, 'Query', df_query['Texto'])

result_df.head(20)
```

```python
# Convertir todas las similitudes a valores numéricos
result_df.iloc[:, 1:] = result_df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# Ordenar las columnas de mayor a menor para cada fila, excluyendo la columna "Query"
sorted_columns = result_df.iloc[:, 1:].apply(lambda row: row.sort_values(ascending=False).index, axis=1)

# Crear un nuevo DataFrame con los nombres de los corpus ordenados
sorted_df = pd.DataFrame({
    'Query': result_df['Query'],
    'Sorted Corpus': sorted_columns.apply(lambda x: ', '.join([f'{col}' for col in x]))
})

sorted_df.head(20)
```
